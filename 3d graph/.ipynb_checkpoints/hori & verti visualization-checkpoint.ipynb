```python
file1 = open('SD_2021.12.23_17.08.59_000_4ms_amp0.log', 'r')
file2 = open('SD_2021.12.23_17.08.59_000_4ms_amp1.log', 'r')
file3 = open('SD_2021.12.23_17.08.59_000_4ms_amp2.log', 'r')
```

```python
lines1 = file1.read().splitlines()
lines2 = file2.read().splitlines()
lines3 = file3.read().splitlines()

file1.close()
file2.close()
file3.close()
```

```python
data1=[]
data2=[]
data3=[]
    
for line in lines1:
    if not line:
        continue

    columns = [col.strip() for col in line.split('\t') if col]
    data1.append(columns)

for line in lines2:
    if not line:
        continue

    columns = [col.strip() for col in line.split('\t') if col]
    data2.append(columns)

for line in lines3:
    if not line:
        continue

    columns = [col.strip() for col in line.split('\t') if col]
    data3.append(columns)
    # do something
```

```python
np_arr1 = np.array(data1, dtype=np.int64)
np_arr2 = np.array(data2, dtype=np.int64)
np_arr3 = np.array(data3, dtype=np.int64)
```

```python
sh_dataset = []
Xh = []
Yh = []
Zh = []

for i in range(len(horizon_arr)):
    for j in range(len(horizon_arr[i])):
        
        sh_dataset.append([i+1, j+1, int(horizon_arr[i][j])])

       
for i in range(len(sh_dataset)):
    Xh.append(sh_dataset[i][0])
    Yh.append(sh_dataset[i][1])
    Zh.append(sh_dataset[i][2])

    
sv_dataset = []
Xv = []
Yv = []
Zv = []

for i in range(len(vertical_arr)):
    for j in range(len(vertical_arr[i])):
        
        sv_dataset.append([i+1, j+1, int(vertical_arr[i][j])])

       
for i in range(len(sv_dataset)):
    Xv.append(sv_dataset[i][0])
    Yv.append(sv_dataset[i][1])
    Zv.append(sv_dataset[i][2])


#s_dataset=np.array(s_dataset)

#s_dataset=s_dataset.reshape(10000,18000,1)
```
